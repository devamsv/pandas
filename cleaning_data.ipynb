```python
import pandas as pd
people = pd.read_csv('people-100.csv')
```

```python
people.isnull() # Check for null values
```

```python
people.notnull() # Check for non-null values
```

```python
people = people.dropna() # Drop rows with any null values
```

```python
people.fillna('Unknown') # Fill null values with 'Unknown'
```

```python
people.replace(1,"one")
```

```python
people.rename(columns={'Email': 'Email_ID'}, inplace=True) # Rename column 'Email' to 'Email_ID'
```

```python
people.drop_duplicates(inplace=True) # Remove duplicate rows
```

```python
people.reset_index(drop=True, inplace=True) # Reset index after dropping rows
```
