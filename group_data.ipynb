```python
import pandas as pd
df = pd.read_csv('BMW_data.csv')
```

```python
df.groupby('Price_USD').sum() # Group the DataFrame by 'Price_USD' and calculate the sum for each group
```

```python
df.head()
```

```python
df.groupby('Sales_Volume').count() # Group the DataFrame by 'Sales_Volume' and calculate the mean for each group
```

```python
df.groupby('Mileage_KM').mean(10) # Group the DataFrame by 'Mileage_KM' and calculate the mean for each group
```

```python
df.agg({'Price_USD': 'max', 'Mileage_KM': 'min'}) # Aggregate to find max Price_USD and min Mileage_KM
```
